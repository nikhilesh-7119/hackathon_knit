```python
import re

# 1 Normalize the newline characters (convert \n to space)
df['State'] = df['State'].astype(str).str.replace(r'\\n', ' ', regex=True)   # handles literal '\n'
df['State'] = df['State'].str.replace(r'\n', ' ', regex=True)                # handles real line breaks

# 2 Replace any 'Multi-State(s)' entry (and everything after) with 'All India'
df['State'] = df['State'].replace(r'(?i)^Multi[\s\-]*States?.*', 'All India', regex=True)

# 3 (Optional) Also clean similar 'All India (...)' entries
df['State'] = df['State'].replace(r'(?i)^All\s*India.*', 'All India', regex=True)
```

```python
# Identify "Total" rows in the Project_Name column
total_rows = df[df['Project_Name'].str.strip().str.lower() == 'total']

# Drop them from main DataFrame (keep only actual projects)
df = df[df['Project_Name'].str.strip().str.lower() != 'total']
```

```python
df['SNo'] = df['SNo'].fillna(method='ffill')
```

```python
df['SNo'].isnull().sum()
```

```python
# Clean and standardize the column
df['State'] = df['State'].astype(str).str.strip()

# Replace any weird NaN-like text or blanks with actual NaN
df['State'] = df['State'].replace(
    [r'(?i)^nan$', r'(?i)^none$', r'(?i)^na$', r'(?i)^null$', r'^\s*$', r'^<na>$'],
    pd.NA,
    regex=True
)

# Normalize line breaks and hidden characters
df['State'] = df['State'].str.replace(r'\\n', ' ', regex=True)
df['State'] = df['State'].str.replace(r'\n', ' ', regex=True)

# Handle 'Multi-State' or 'All India (XYZ)' type text
df['State'] = df['State'].replace(r'(?i)^Multi[\s\-]*States?.*', 'All India', regex=True)
df['State'] = df['State'].replace(r'(?i)^All\s*India.*', 'All India', regex=True)

#  Conditional fill  detect national projects from Project_Name
mask_national = df['Project_Name'].str.contains(
    r'\b(National|Pan\s*India|Corridor|Expressway|Highway|Bharatmala|Sagarmala)\b',
    case=False,
    na=False
)
df.loc[mask_national, 'State'] = 'All India'

#  Finally, fill remaining blanks or nulls
df['State'] = df['State'].fillna('Unknown')
```

```python
df['State'].value_counts()
```

```python
# Step 1  Convert Expenditure to numeric (if needed)
df['Expenditure'] = pd.to_numeric(df['Expenditure'], errors='coerce')

# Step 2  Group by Agency (or change 'Agency' to your sector column)
sector_means = df.groupby('Agency')['Expenditure'].mean()

# Step 3  Fill missing Expenditure with that sectors mean
df['Expenditure'] = df.apply(
    lambda row: sector_means[row['Agency']] if pd.isna(row['Expenditure']) and row['Agency'] in sector_means else row['Expenditure'],
    axis=1
)

# Step 4  If some sectors had all NaN (so still missing), fill with global mean
df['Expenditure'] = df['Expenditure'].fillna(df['Expenditure'].mean())
```

```python
df['Expenditure'].value_counts()
```

```python
df['Progress'] = pd.to_numeric(df['Progress'], errors='coerce')  # convert non-numeric to NaN
df['Progress'] = df['Progress'].fillna(0)
```

```python
df['Agency'] = df['Agency'].fillna('Unknown Agency')
df['Project_Name'] = df.apply(
    lambda row: f"Unnamed Project - {row['Agency']}" if pd.isna(row['Project_Name']) else row['Project_Name'],
    axis=1
)
```

```python
df['Dateof_Approval'] = df['Dateof_Approval'].fillna('NA')
df['Start_Date'] = df['Start_Date'].fillna('NA')
```

```python
df['Org_DoC'] = df['Org_DoC'].fillna('NA')
df['Rev_DoC'] = df['Rev_DoC'].fillna(df['Org_DoC'])
```

```python
# 1 Ensure numeric columns
df['Org_Cost'] = pd.to_numeric(df['Org_Cost'], errors='coerce')
df['Rev_Cost'] = pd.to_numeric(df['Rev_Cost'], errors='coerce')

# 2 Compute median Org_Cost and Rev_Cost per Agency
agency_org_median = df.groupby('Agency')['Org_Cost'].median()
agency_rev_median = df.groupby('Agency')['Rev_Cost'].median()

# 3 Fill missing Org_Cost sector-wise
df['Org_Cost'] = df.apply(
    lambda row: agency_org_median[row['Agency']]
    if pd.isna(row['Org_Cost']) and row['Agency'] in agency_org_median
    else row['Org_Cost'],
    axis=1
)

# 4 Fill missing Rev_Cost sector-wise
df['Rev_Cost'] = df.apply(
    lambda row: agency_rev_median[row['Agency']]
    if pd.isna(row['Rev_Cost']) and row['Agency'] in agency_rev_median
    else row['Rev_Cost'],
    axis=1
)

# 5 Fill any still-missing Org_Cost with global median
df['Org_Cost'] = df['Org_Cost'].fillna(df['Org_Cost'].median())

# 6 Fill any still-missing Rev_Cost with corresponding Org_Cost
df['Rev_Cost'] = df['Rev_Cost'].fillna(df['Org_Cost'])
```
